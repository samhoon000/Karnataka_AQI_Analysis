# Getting the dataset from kaggle ( Dataset=India Air Quality Data)

```python
import kagglehub
path = kagglehub.dataset_download("shrutibhargava94/india-air-quality-data")
```

```python
import pandas as pd
import os

print(os.listdir(path))

csv_file_name = 'data.csv'
full_csv_path = os.path.join(path, csv_file_name)

df = pd.read_csv(full_csv_path, encoding='latin1')
display(df.head())
```

```python
df.info()
```

```python
df=df.drop(columns=['stn_code','agency','location_monitoring_station','type','spm','sampling_date'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['date']=pd.to_datetime(df['date'],errors='coerce')
```

```python
df['state'].unique()
```

```python
ka=df[df['state']=='Karnataka'].copy()
```

```python
ka['location'].unique()
```

```python
def get_season(month):
  if month in [11,12,1,2]:
    return 'Winter'
  elif month in [3,4,5]:
    return 'Summer'
  else:
    return "Monsoon"
```

```python
ka['season']=ka['date'].dt.month.apply(get_season)
```

```python
ka.head()
```

```python
ka.info()
```

```python
ka=ka.drop(columns=['pm2_5'])
```

```python
ka.head()
```

# The dataset is ready for analysis

```python
ka.to_csv('karnataka_air_quality.csv',index=False)
```
